```python
import numpy as np
import pandas as pd

pm=open('purepm_98_TMR.fasta','r')

fas = [[]]

while True:
    reading = pm.readline()
    reading = reading.rstrip()

    if not reading.startswith('>'):
        if reading == '':
            break
        fas.append(list(reading))

fas.remove([])
columns=list(range(-25,26))
df_pm=pd.DataFrame(fas,columns=columns)

df_pm
```

```python
npm=open('npm_98_TMR.fasta','r')

fas = [[]]

while True:
    reading = npm.readline()
    reading = reading.rstrip()

    if not reading.startswith('>'):
        if reading == '':
            break
        fas.append(list(reading))

fas.remove([])
columns=list(range(-25,26))
df_npm=pd.DataFrame(fas,columns=columns)

df_npm
```

```python
L_list=[]

for index_pm, row_pm in df_pm.iterrows():
    i=0
    L_count=0
    while i<5:
        if row_pm[4+i]=='L':
            L_count=L_count+1
        i=i+1
    L_list.append(L_count)
#for index_npm, row_npm in df_npm.iterrows():

i=0
count_list=[0,0,0,0,0,0]

while i<len(L_list):
    if L_list[i]==0:
        count_list[0]=count_list[0]+1
    elif L_list[i]==1:
        count_list[1]=count_list[1]+1
    elif L_list[i]==2:
        count_list[2]=count_list[2]+1
    elif L_list[i]==3:
        count_list[3]=count_list[3]+1
    elif L_list[i]==4:
        count_list[4]=count_list[4]+1
    else:
        count_list[5]=count_list[5]+1        
    i=i+1
    
import numpy as np
import matplotlib.pyplot as plt

print ("L_pm")
x = np.array(count_list)
label = ["0", "1", "2", "3", "4",""]
plt.pie(x,labels=label,counterclock=False, startangle=90)
plt.show

print (count_list)

```
